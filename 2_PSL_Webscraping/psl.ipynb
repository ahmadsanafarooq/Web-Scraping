```python
import requests
from bs4 import BeautifulSoup

url = "https://psl-t20.com/psl-2025-points-table/"
headers = {
    "User-Agent": "Mozilla/5.0"
}

response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, "html.parser")

# Look for all table tags
tables = soup.find_all("table")
print(f"Tables found: {len(tables)}")
```

```python
import pandas as pd

url=("https://en.wikipedia.org/wiki/Pakistan_Super_League")
tables=pd.read_html(url)
#check how many tables were found
print(f"Number of tables found: {len(tables)}")
# Display the first table
print(tables[0].head())
```

```python
# print all the tables
for i, table in enumerate(tables):
    print(f"Table {i}:")
    print(table.head())
    print("\n")
```

```python
# save all tables to csv files
for table in tables:
    table_name = table.columns[0] if not table.empty else "table"
    table.to_csv(f"{table_name}.csv", index=False)
    print(f"Saved {table_name}.csv")
```

```python
#combine all tables into a single DataFrame
combined_df = pd.concat(tables, ignore_index=True)
# save the combined DataFrame to a csv file
combined_df.to_csv("combined_tables06_23_2025.csv", index=False)
```
