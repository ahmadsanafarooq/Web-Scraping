```python
import pandas as pd
```

```python
# Load the URL for world population on wikipedia
url = "https://en.wikipedia.org/wiki/List_of_countries_and_dependencies_by_population"
tables=pd.read_html(url)
#check how many tables were found
print(f"Number of tables found: {len(tables)}")
# Display the first table
print(tables[0].head())
```

```python
#show all tables
for i, table in enumerate(tables):
    print(f"Table {i}:")
    print(table.head(), "\n")
```

```python
# step 1: Import the required libraries
import pandas as pd

# step 2: Load the URL
url = 'https://www.worldometers.info/world-population/population-by-country/'

# step 3: Fetch the HTML content with headers and read the HTML table into a DataFrame
import requests
# headers to avoid being blocked
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'}
# headers are important to avoid being blocked by the server
response = requests.get(url, headers=headers)
response.raise_for_status()  # Raise an error for bad status codes

tables = pd.read_html(response.text)

# step 4: Check the number of tables
print(f"Number of tables found: {len(tables)}")
# step 5: Display the first few rows of the first table
tables[0].head()
```

```python
# Save the first table to a CSV file
tables[0].to_csv('world_population_by_country.csv', index=False)
```
